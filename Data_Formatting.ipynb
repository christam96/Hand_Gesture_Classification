```python
# Imports
import pandas as pd
import numpy as np
```

```python
df_0 = pd.read_csv('EMG Data/0.csv')
df_1 = pd.read_csv('EMG Data/1.csv')
df_2 = pd.read_csv('EMG Data/2.csv')
df_3 = pd.read_csv('EMG Data/3.csv')
```

## Formatting Gesture 0

```python
df_0.head()
```

```python
print('Original shape: ', df_0.shape)
df_0.drop(df_0.columns[len(df_0.columns)-1], axis=1, inplace=True)
print('Shape after removing gesture column: ', df_0.shape)
```

```python
bigdata_0 = pd.DataFrame()
for row in df_0.iloc[:,].values:
    reshape = np.reshape(row, (8,8))
    temp_df = pd.DataFrame(reshape)
    bigdata_0 = bigdata_0.append(temp_df,ignore_index=True)

print('Shape after reformatting: ', bigdata_0.shape)
print('Reformatted dataframe (no gesture column): ')
bigdata_0
```

```python
# Output first 24 rows
bigdata_0.iloc[:24,:]
```

```python
# Add gesture column
bigdata_0['Gesture'] = 0
print('\nReformatted dataframe:')
bigdata_0
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))
plt.plot(bigdata_0[5])
```
